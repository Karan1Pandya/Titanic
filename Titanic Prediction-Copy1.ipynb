```python
# x--Features
x=data.drop('Survived',axis=1)
```

```python
# y---target variable----its the column that is to be pridected using ml
y=data['Survived']
```

```python
x.drop(['PassengerId'],axis=1,inplace=True)
x.drop(['Ticket'],axis=1,inplace=True)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=7)
# 0.7 is 70% data when we write random_state=7 every time it will take same random data fow whole team 7 can be any no.
# train_test_split() returns a list of 4 dataframes
```

```python
rf=RandomForestClassifier()
```

```python
rf.fit(x_train.drop('Name',axis=1),y_train)
```

```python
y_pred=rf.predict(x_test.drop('Name',axis=1))
```

```python
x_test['pred_survived']=y_pred
```

```python
x_test['act_surv']=y_test
```
